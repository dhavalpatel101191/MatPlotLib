```python
import numpy as np
from matplotlib import pyplot as plt
import os
import pandas as pd
```

```python
pwd = os.getcwd()

File = pd.read_csv(pwd + '/data.csv')
#File
```

#.......Line Chart........#

```python
# Set the value for X and Y

x = np.arange(1,11)
y = 2*x
```

```python
# Lable the Axis and Title of Chart

plt.plot(x,y)
plt.title('Line Chart')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()                    # .show() method to show the chart
```

```python
plt.plot(x,y,color='Orange',linestyle=':',linewidth=5)
plt.title('Line Chart')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show() 
```

```python
y1 = 2*x
y2 = 3*x
```

```python
plt.plot(x,y1,color='Orange',linestyle=':',linewidth=2)
plt.plot(x,y2,color='Green',linestyle='-.',linewidth=2)
plt.title('Line Chart')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(True)         # Grid show in Chart
plt.show() 
```

```python
# Adding Sub Plot
# .subplot(Row,Column,Index)
plt.subplot(2,1,1)
plt.plot(x,y1,color='Orange',linestyle=':',linewidth=2)
plt.title('Line Chart 1')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.subplot(2,1,2)
plt.plot(x,y2,color='Green',linestyle='-.',linewidth=2)
plt.title('Line Chart 2')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

#........Bar Plot.......#

```python
student = {'Bob':90,'Matt':70,'Simon':50}
Name = list(student.keys())
Marks = list(student.values())


# Verticle Bar Chart
#.bar() method
plt.bar(Name,Marks,color ='Green')

plt.title('Bar Graph')
plt.xlabel('Name')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

```python
# Horizontal Bar Chart
# barh() method
plt.barh(Name,Marks,color ='Red')

plt.title('Bar Graph')
plt.xlabel('Name')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

```python
# Scatter Plot
# .scatter()  method

x = list(range(10,100,10))
y = [8,5,7,6,3,2,8,7,5]

plt.scatter(x,y)
plt.show()
```

```python
# Use different Indicator and Color
plt.scatter(x,y,marker='*',color = 'Red',s= 100)
plt.show()
```

```python
# Two scatter plot in one
z = [5,2,3,6,4,1,8,7,9]
plt.scatter(x,y,marker='*',color = 'red',s=100)
plt.scatter(x,z,marker='.',color='Green',s=100)
plt.show()
```

```python
# Subplot of Scatter plot
# .subplot() method

plt.subplot(2,1,1)
plt.scatter(x,y,marker='*',color = 'red',s=100)
plt.show()
plt.subplot(2,1,2)
plt.scatter(x,z,marker='.',color='Green',s=100)
plt.show()
```

#..........Histogram........# (For Numerical Value)

```python
# Histogram
# .hist()  method

data = [8,5,6,9,8,7,5,2,6,5,6,8,5,4]
plt.hist(data)
plt.show()
```

```python
plt.hist(data,color='Red',bins=5)
plt.grid(True)
plt.show()
```

```python
plt.hist(File['Maxpulse'],color='Green',bins=30)
plt.show()
```

#......Box-Plot.......#

```python
# Box Plot  
# .boxplot() method

one = list(range(1,10))
two = [1,2,3,4,5,4,3,2,1]
three = [6,7,8,9,8,7,6,5,4]

box = list([one,two,three])

plt.boxplot(box)
plt.show()
```

```python
data1 = list([File['Pulse'],File['Maxpulse'],File['Duration']])
plt.boxplot(data1)
plt.show()
```

#.....PieChart.....#

```python
# pie() method
#pie(Numrical value,labels = categorical value)

fruit = ['Apple','Orange','Mango','Guava']
qty = [10,50,80,60]

plt.pie(qty,labels = fruit)
plt.show()
```

```python
plt.pie(qty,labels = fruit,autopct='%0.1f%%',colors=['Yellow','White','Pink','Red'])
plt.show()
```

#......DoughNut Chart......#

```python
plt.pie(qty,labels=fruit,radius=2)
plt.pie([1],colors=['White'],radius=1)
plt.show()
```
